For our project we chose to work with Brave Bar Co. which is a protein bar company that was started by a fellow CU student. The hope of the founder was to create a healthy protein bar made from simple, raw, locally sourced ingredients. As the owner of the company is still a student and the company is still young, he does a majority of his advertising on social media and specifically, on instagram. Instagram provides a number of metrics for its users that give details on how much interaction posts are generating with the account. For our project we set out to take the data provided by the app and put into a more user friendly format such that the owner would be able to make more strategic marketing decisions based off of that information.




The code below simply presents the input data that we recieved from the owner of Brave Bar Co. 


```python
from pandas.io.formats.info import DataFrameInfo
import pandas as pd


DataFrameInfo = pd.read_excel("bravebar1.xlsx")
df1 = DataFrameInfo.fillna(0)
df1.rename(columns={'Impression': 'Total Impressions'}, inplace=True)


df1
```

This code Groups the various rows by the type of advertising, then takes a sum of the impressions and ranks them to give the business owner a better idea of which types of advertising are most effective 

```python
df2 = df1.groupby('Type').sum()

dfrank = df2['Total Impressions'].rank()
dfrank
```
